## Exercise 1: Sample behaviour of an MDP

Let's take (again) a look at Sutton & Barto example 4.1 gridworld. On each iteration start at every (non-terminating) state and sample actions in succeeding states by selecting them from uniform distribution (each action - up, down, left, right - is equally probable). Run the episode until terminal state is encountered. Collect statistics to calculate average number of steps needed before completion for each start state. Should this number match with something you have seen earlier in the exercises?

```python
# YOUR CODE
rows_count = 4
columns_count = 4
V = np.zeros((rows_count, columns_count))
terminating = [(0,0), (rows_count-1, columns_count-1)]

def valueFor(row, column, direction):
    if (direction == 'up'):
        if (row == 0):
            r1 = 0
            c1 = column
            
        else:
            r1 = row -1
            c1 = column
    if (direction == 'left'):
        if (column == 0):
            r1 = row
            c1 = 0
        else:
            r1 = row
            c1 = column -1
    if (direction == 'down'):
        if (row == 3):
            r1 = row
            c1 = column
        else:
            r1 = row +1
            c1 = column
    
    if (direction == 'right'):
        if (column == 3):
            r1 = row
            c1 = column
        else:
            r1 = row
            c1 = column + 1

    return (r1,c1)


for i in range(1000):
    for r in range(rows_count):
        for c in range (columns_count):
            row=r
            column=c
            steps=0
            while not (row, column) in terminating:

                x = randint(1, 4)

                if (x == 1):
                    a=valueFor(row,column, 'up')
                elif (x==2):
                    a= valueFor(row,column, 'down')
                elif (x==3):
                    a=valueFor(row,column, 'left')

                else:
                    a= valueFor(row,column, 'right')

                row = a[0]
                column = a[1]
                steps += 1
                
         # m= m + (steps-m)/y   
            V[r,c] += steps
    

with np.printoptions(precision=2):
    print(V/1000) 
```
